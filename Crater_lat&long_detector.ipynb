
This is a code for detecting the latitude and longitude of all the lunar craters. This is working with crater dataset. The dataset contains the metadata of 291286 images from LROC database. The x and y of craters can be manually insert to the code. The code should be updated by adding the craters picxel dataset. The crater picxel dataset contains the picxel(x,y) of the crater's center in each image. It would be Provided by the CDA program.  

```python
#importing the libraries 
import pandas as pd
import numpy as np

#Read the .txt metadata dataset 
df=pd.read_csv(r'C:\Users\s\Desktop\Capstone\meta data\crater.txt',delimiter=",")
#df.head(3)

#Reading the columns name
# for col in df.columns: 
#     print(col) 




```

```python
#Crating the img datafarame from the actual dataset 
imgcor = df[['Image Name','Upper left latitude','Upper left longitude','Lower left latitude','Upper right longitude','Image lines','Line samples']]

#Manually creating the crater x and y with zero input. This part should be updated with CDA program result dataset
imgcor['crater x']= 0
imgcor['crater y']= 0




imgcor
```

```python
#Calculating the latitude of crater. yc should be replaced by the from imgcor data farme 'crater y' 
#Insert the crater y
yc = 0
a_lat = (imgcor['Upper left latitude'] - imgcor['Lower left latitude']) / (0 - imgcor['Image lines'])
d_lat = (yc-0)

latc = a_lat * d_lat + imgcor['Upper left latitude']


#Calculating the longitude of crater. xc should be replaced by the from imgcor data farme 'crater x' 
#insert the crater x
xc = 0
a_long = (imgcor['Upper left longitude'] - imgcor['Upper right longitude']) / (0 - imgcor['Line samples'])
d_long = (xc-0)

longc = a_long * d_long + imgcor['Upper left longitude']

#print the crater llatitude and longitude
# print(' crater lat=', latc)
# print(' crater long =', longc)


# add the lat and long to data frame.

imgcor['crater latitude']= latc
imgcor['crater longitude']=longc



```
