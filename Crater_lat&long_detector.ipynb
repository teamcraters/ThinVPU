
This is a code for detecting the latitude and longitude of all the lunar craters. It is working with crater dataset. The crater dataset contains the metadata of 291286 images from LROC database. The x and y of craters can be manually insert to the code. The code should be updated by adding the craters picxel dataset. The crater picxel dataset contains the picxel(x,y) of the crater's center in each image. It would be Provided by the CDA program.  

```python
#importing the libraries 
import pandas as pd
import numpy as np

#Read the .txt metadata dataset 
df=pd.read_csv(r'C:\Users\s\Desktop\Capstone\meta data\crater.txt',delimiter=",")
df.head(3)

#Reading the columns name
for col in df.columns: 
    print(col) 
```

```python
#Crating the corimg datafarame from the actual dataset 
corimg = df[['Image Name','Upper left latitude','Upper left longitude','Lower left latitude','Upper right longitude','Lower left longitude','Image lines','Line samples']]

corimg.head(3)
```

```python
#Calculating the latitude of crater. yc should be replaced by the from imgcor data farme 'crater y' 
#Insert the crater y
yc = 41917
s_lat = (corimg['Upper left latitude'] - corimg['Lower left latitude']) / (0 - corimg['Image lines'])
d_lat = (yc-0)

latc = s_lat * d_lat + corimg['Upper left latitude']


#Calculating the longitude of crater. xc should be replaced by the from imgcor data farme 'crater x' 
#insert the crater x
xc = 502
s_long = (corimg['Upper left longitude'] - corimg['Upper right longitude']) / (39 -corimg['Line samples'])
d_long = (xc-39)

longc = s_long * d_long + corimg['Lower left longitude']

#print the crater llatitude and longitude
# print(' crater lat=', latc)
# print(' crater long =', longc)


# add the lat and long to data frame.
corimg['crater latitude']= latc
corimg['crater longitude']=longc


corimg


```
